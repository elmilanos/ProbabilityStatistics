## Problema 1: Distribución Binomial - Accidentes de tránsito en una intersección
La distribución **Binomial** es adecuada porque estamos analizando eventos independientes en un intervalo fijo de 10 minutos, donde cada minuto tiene una probabilidad de accidente de 0.05.

Los parámetros son:

- **n = 10** (minutos analizados).

- **p = 0.05** (probabilidad de accidente por minuto).

- **k** (número de accidentes en 10 minutos).

Usaremos *scipy.stats.binom* para generar simulaciones y graficar la distribución.

```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as stats

# --- PROBLEMA 1: DISTRIBUCIÓN BINOMIAL (Accidentes de tránsito) ---
n = 10  # Número de minutos
p = 0.05  # Probabilidad de accidente por minuto
k_values = np.arange(0, 11)  # Posibles valores de accidentes (0 a 10)

# Simulación de 1000 intersecciones de 10 minutos
simulaciones_binomial = np.random.binomial(n, p, 1000)

# Gráfica de la distribución binomial
plt.figure(figsize=(10, 5))
plt.hist(simulaciones_binomial, bins=np.arange(-0.5, 10.5, 1), density=True, alpha=0.6, color='b', edgecolor='black', label="Simulación")
plt.plot(k_values, stats.binom.pmf(k_values, n, p), 'ro-', label="Distribución Teórica Binomial")
plt.xlabel("Número de accidentes en 10 minutos")
plt.ylabel("Probabilidad")
plt.title("Distribución Binomial - Accidentes en 10 minutos")
plt.legend()
plt.grid()
plt.show()

# Cálculo del valor esperado y varianza
media_binomial = n * p
varianza_binomial = n * p * (1 - p)

print(f"Valor esperado (media) de accidentes: {media_binomial}")
print(f"Varianza de accidentes: {varianza_binomial}")
```

## Problema 2: Distribución Poisson - Llegada de vehículos al centro de Durango
La distribución Poisson es adecuada porque modela eventos que ocurren en un intervalo continuo (llegada de vehículos).

Los parámetros son:

- **λ = 5** (promedio de vehículos por minuto).

- **Tiempo total = 15 minutos** (λ total = 5 * 15 = 75).

- Modelamos la llegada de autos como eventos independientes en cada minuto.

Nuevamente usaremos *scipy.stats.poisson* para simular y graficar la distribución.

```python
# --- PROBLEMA 2: DISTRIBUCIÓN POISSON (Llegada de vehículos) ---
lambda_minuto = 5  # Tasa de llegada de autos por minuto
lambda_total = lambda_minuto * 15  # Para 15 minutos

# Simulación de llegadas de autos en 1000 intervalos de 15 minutos
simulaciones_poisson = np.random.poisson(lambda_total, 1000)

# Gráfica de la distribución Poisson
plt.figure(figsize=(10, 5))
plt.hist(simulaciones_poisson, bins=np.arange(min(simulaciones_poisson)-0.5, max(simulaciones_poisson)+0.5, 1), density=True, alpha=0.6, color='g', edgecolor='black', label="Simulación")
x_poisson = np.arange(40, 110)
plt.plot(x_poisson, stats.poisson.pmf(x_poisson, lambda_total), 'ro-', label="Distribución Teórica Poisson")
plt.xlabel("Número de vehículos en 15 minutos")
plt.ylabel("Probabilidad")
plt.title("Distribución Poisson - Llegada de vehículos en 15 minutos")
plt.legend()
plt.grid()
plt.show()

# Cálculo del valor esperado y varianza
media_poisson = lambda_total
varianza_poisson = lambda_total

print(f"Valor esperado (media) de vehículos en 15 minutos: {media_poisson}")
print(f"Varianza de vehículos: {varianza_poisson}")
```
