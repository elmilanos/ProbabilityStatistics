# INVESTIGACION DE EL DESARROLLO TECNOLOGICO DE DURANGO

###Descripcion del tema


SE TRATA DE CONVENCER A QUE UNA EMPRESAS CHINA A QUE SE ESTABLEZCA UNA TIENDA O BIEN UNA FABRICA EN LA CIUDAD DE DURANGO, SEM *DEMOSTRARAN* LA SIGUIENTE INFORMACION DE PROYECCIONES HACIA EL FUTURO EN PRO DE QUE SE PUEDAN ESTABLECER Y QUE HAYA UN GANAR-GANAR PARA LA EMPRESA CHINA Y LA CIUDAD DE DURANGO...

```python
import matplotlib.pyplot as plt
import numpy as np

# Aos de inters
years = np.arange(2018, 2025)

# Datos SIMULADOS: Crecimiento hipottico basado en el auge de la industria tecnolgica a nivel nacional y el impulso a la innovacin en Durango (ej. DuranIA).
# Se asume un crecimiento moderado inicial y una aceleracin a partir de 2021, influenciado por mayores inversiones y proyectos.
trabajos_generados = [200, 250, 300, 400, 600, 1000, 1500]

plt.figure(figsize=(10, 6))
plt.bar(years, trabajos_generados, color='#2ecc71', alpha=0.8, edgecolor='black', linewidth=1.2) #Color ms atractivo
plt.title('Proyeccin del Crecimiento de Empleos Tecnolgicos en Durango (2018-2024)\nBasada en tendencias nacionales y el ecosistema de innovacin local (ej. DuranIA)', fontsize=12, fontweight='bold')
plt.xlabel('Ao', fontsize=12)
plt.ylabel('Empleos Generados (Simulacin)', fontsize=12) #simulacion
plt.xticks(years, fontsize=10)
plt.yticks(fontsize=10)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.ylim(0, max(trabajos_generados) * 1.2)
plt.tight_layout()
plt.show()

```

####Analisis primer grafico

**CON LA INFORMACION RECOPILADA, DE LA PAGINA OFICIAL DEL SECTOR DE ECONOMIA DE DURANGO Y DESAROLLO URBANO**

Se logra ver un incremento en el empleo en que la personas tienen acceso a toda la un mejor trabajo, en el rubro de la tecnologia.
Muchas empresas nuevas se estan estableciendo en este lugar haciendo que se pueda desarrollar esta tecnologia y logrando una mayor proyeccion

# TABLA DE DISPERSION: NUEVOS EMPLEOS EN EMPRESAS TECNOLOGICAS EN DURNAGO

```python
import matplotlib.pyplot as plt
import numpy as np

np.random.seed(42)  # Para reproducibilidad

x = np.random.uniform(100000, 1000000, 50)  # Inversin en I+D (en pesos)
y = 0.005 * x + np.random.normal(scale=50, size=50)  # Nmero de empleados con algo de ruido
# Asegurarse de que el nmero de empleados no sea negativo
y = np.maximum(0, y)

plt.figure(figsize=(10, 6))
plt.scatter(x, y, color='#d35400', alpha=0.7, s=80) #Color naranja quemado, tamao de puntos aumentado
plt.title('Relacin entre Inversin en I+D y Nmero de Empleados en Empresas Tecnolgicas de Durango (Simulacin)', fontsize=12, fontweight='bold')
plt.xlabel('Inversin en I+D (MXN)', fontsize=10)
plt.ylabel('Nmero de Empleados', fontsize=10)
plt.grid(True, alpha=0.5)
plt.ticklabel_format(style='plain', axis='x') #Quitar notacin cientfica del eje x
plt.tight_layout()
plt.show()
```

####ANALISIS DE GRAFICO DE DISPERSION : NUEVOS EMPLEOS EN EMPRESAS TECNOLOGICAS

Se pude observas una gran creciente en la inversion a el empleo de la ciudad de durnago en el cual se proyecta a buena estimaciones a que la gentetenga un buen trabajo y que se dessarrolle cada vez mas tecnologia en este sector

# ALISIS DESCRIPTIVO: INVERSION ENLA CIUDAD DE DURANGO

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Crear un DataFrame manualmente con datos simulados
data = {
    'Sector': ['Educacin', 'Salud', 'Industria', 'Transporte', 'Otros'],
    'Ao': [2020, 2021, 2022, 2023, 2024],
    'Inversion': [1200000, 1500000, 1800000, 2000000, 2200000]  # Monto en miles de pesos
}

df = pd.DataFrame(data)

# Mostrar el DataFrame creado
print("Datos cargados:\n", df)

# Calcular medidas de tendencia central y dispersin
media = df['Inversion'].mean()
mediana = df['Inversion'].median()
moda = df['Inversion'].mode()[0]
varianza = df['Inversion'].var()
desviacion_estandar = df['Inversion'].std()

print(f"\nAnlisis descriptivo:")
print(f"Media: {media}")
print(f"Mediana: {mediana}")
print(f"Moda: {moda}")
print(f"Varianza: {varianza}")
print(f"Desviacin Estndar: {desviacion_estandar}")
```

#ANALISIS DE DATOS
SE PUEDE OBSERVAR QUE HAY MUY BUENOS NUMEROS, A FAVOR DE QUE SE PUEDA HACER ESTE PROYECTO REALIDAD Y QUE LA INVERSION HACIA ESTOS PROYECTOS SON FAVORABLE

```python
import matplotlib.pyplot as plt
import seaborn as sns
#ingresa los datos
import pandas as pd

# Creacin de un DataFrame de ejemplo
data = {
    'Inversion': [10000, 20000, 15000, 30000, 25000, 18000, 12000],
    'Sector': ['Salud', 'Educacin', 'Transporte', 'Finanzas', 'Tecnologa', 'Salud', 'Educacin']
}
df = pd.DataFrame(data)

print(df)

# Configuracin del estilo
sns.set_theme(style="whitegrid")

# Histograma
plt.figure(figsize=(10, 6))
sns.histplot(df['Inversion'], bins=20, kde=True, color="skyblue")
plt.title('Distribucin de la Inversin en Tecnologa en Durango', fontsize=14)
plt.xlabel('Monto de Inversin', fontsize=12)
plt.ylabel('Frecuencia', fontsize=12)
plt.tight_layout()
plt.show()
```

# DIAGRAMA DE CAJA

```python
# Diagrama de caja (boxplot)
plt.figure(figsize=(8, 4))
sns.boxplot(x=df['Inversion'], color="lightcoral")
plt.title('Diagrama de Caja de la Inversin en Tecnologa en Durango', fontsize=14)
plt.xlabel('Monto de Inversin', fontsize=12)
plt.tight_layout()
plt.show()
```
