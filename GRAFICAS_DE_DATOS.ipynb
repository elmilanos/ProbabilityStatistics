INVESTIGACION DE EL DESARROLLO TECNOLOGICO DE DURANGO

```python
import matplotlib.pyplot as plt
import numpy as np

# Años de interés
years = np.arange(2018, 2025)

# Datos SIMULADOS: Crecimiento hipotético basado en el auge de la industria tecnológica a nivel nacional y el impulso a la innovación en Durango (ej. DuranIA).
# Se asume un crecimiento moderado inicial y una aceleración a partir de 2021, influenciado por mayores inversiones y proyectos.
trabajos_generados = [200, 250, 300, 400, 600, 1000, 1500]

plt.figure(figsize=(10, 6))
plt.bar(years, trabajos_generados, color='#2ecc71', alpha=0.8, edgecolor='black', linewidth=1.2) #Color más atractivo
plt.title('Proyección del Crecimiento de Empleos Tecnológicos en Durango (2018-2024)\nSimulación basada en tendencias nacionales y el ecosistema de innovación local (ej. DuranIA)', fontsize=12, fontweight='bold')
plt.xlabel('Año', fontsize=12)
plt.ylabel('Empleos Generados (Simulación)', fontsize=12) #simulacion
plt.xticks(years, fontsize=10)
plt.yticks(fontsize=10)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.ylim(0, max(trabajos_generados) * 1.2)
plt.tight_layout()
plt.show()

```

# TABLA DE DISPERSION: NUEVOS EMPLEOS EN EMPRESAS TECNOLOGICAS EN DURNAGO

```python
import matplotlib.pyplot as plt
import numpy as np

np.random.seed(42)  # Para reproducibilidad

x = np.random.uniform(100000, 1000000, 50)  # Inversión en I+D (en pesos)
y = 0.005 * x + np.random.normal(scale=50, size=50)  # Número de empleados con algo de ruido
# Asegurarse de que el número de empleados no sea negativo
y = np.maximum(0, y)

plt.figure(figsize=(10, 6))
plt.scatter(x, y, color='#d35400', alpha=0.7, s=80) #Color naranja quemado, tamaño de puntos aumentado
plt.title('Relación entre Inversión en I+D y Número de Empleados en Empresas Tecnológicas de Durango (Simulación)', fontsize=12, fontweight='bold')
plt.xlabel('Inversión en I+D (MXN)', fontsize=10)
plt.ylabel('Número de Empleados', fontsize=10)
plt.grid(True, alpha=0.5)
plt.ticklabel_format(style='plain', axis='x') #Quitar notación científica del eje x
plt.tight_layout()
plt.show()
```

# ANALISIS DE GRAFICO DE DISPERSION : NUEVOS EMPLEOS EN EMPRESAS TECNOLOGICAS

Se pude observas una gran creciente en la inversion a el empleo de la ciudad de durnago en el cual se proyecta a buena estimaciones a que la gentetenga un buen trabajo y que se dessarrolle cada vez mas tecnologia en este sector

# ALISIS DESCRIPTIVO: INVERSION ENLA CIUDAD DE DURANGO

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Crear un DataFrame manualmente con datos simulados
data = {
    'Sector': ['Educación', 'Salud', 'Industria', 'Transporte', 'Otros'],
    'Año': [2020, 2021, 2022, 2023, 2024],
    'Inversion': [1200000, 1500000, 1800000, 2000000, 2200000]  # Monto en miles de pesos
}

df = pd.DataFrame(data)

# Mostrar el DataFrame creado
print("Datos cargados:\n", df)

# Calcular medidas de tendencia central y dispersión
media = df['Inversion'].mean()
mediana = df['Inversion'].median()
moda = df['Inversion'].mode()[0]
varianza = df['Inversion'].var()
desviacion_estandar = df['Inversion'].std()

print(f"\nAnálisis descriptivo:")
print(f"Media: {media}")
print(f"Mediana: {mediana}")
print(f"Moda: {moda}")
print(f"Varianza: {varianza}")
print(f"Desviación Estándar: {desviacion_estandar}")
```

ANALISIS DESCRIPTIVO DE LA

Se puede observar una mejor espectativa a la proyeccion futura de estas inversiones, ya que se tiene una media o varianza bastante a favor y logrando asi que se convezcan en venir a establecer una tienda.

```python
# Configurar el estilo de los gráficos
sns.set(style="whitegrid")

# Histograma
plt.figure(figsize=(10, 6))
sns.histplot(df['Inversion'], bins=20, kde=True)
plt.title('Distribución de la Inversión en Tecnología en Durango')
plt.xlabel('Monto de Inversión')
plt.ylabel('Frecuencia')
plt.show()

# Diagrama de caja (boxplot)
plt.figure(figsize=(8, 4))
sns.boxplot(x=df['Inversion'])
plt.title('Diagrama de Caja de la Inversión en Tecnología en Durango')
plt.xlabel('Monto de Inversión')
plt.show()

# Gráfico de barras por sector (si se dispone de esta información)
plt.figure(figsize=(12, 6))
sns.barplot(x='Sector', y='Inversion', data=df, estimator=sum)
plt.title('Inversión Total en Tecnología por Sector en Durango')
plt.xlabel('Sector')
plt.ylabel('Inversión Total')
plt.xticks(rotation=45)
plt.show()
```
